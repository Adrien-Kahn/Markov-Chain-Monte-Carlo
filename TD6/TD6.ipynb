# Thermodynamic Integration

```python
N = 10

v_distribution = abs((N + 1) / 2 - np.array(range(1, N+1)))
plt.plot(range(1, N+1), v_distribution)
plt.title("V-shaped distribution")

print("Integration constant: \t{}".format(np.sum(v_distribution)))
print("N^2 / 4: \t\t{}".format(N**2 / 4))
```

```python
def alpha_v_mcmc(pos_ini, n_steps, n, alpha, alpha_dif):
    assert 1 - n%2,"n should be even"
    unnormalized_alpha_v_distribution = [0] + [abs((n + 1)/2 - k)**alpha for k in range(1, n+1)] + [0]
    pos = pos_ini
    observable = 0
    for k in range(n_steps):
        x = np.random.random()
        direction = 2 * np.random.randint(2) - 1
        if x < unnormalized_alpha_v_distribution[pos+direction] / unnormalized_alpha_v_distribution[pos]:
            pos += direction
        observable += abs(((n + 1)/2) - pos)**alpha_dif
    return observable / n_steps
```

```python
n = 10
pos_ini = 1
n_steps = 1000000

# Number of calculation steps for the partition function
n_calculation_steps = 10

res = n
for k in range(n_calculation_steps):
    res *= alpha_v_mcmc(pos_ini, n_steps, n, k/n_calculation_steps, 1/n_calculation_steps)

print(res)
print(n**2 / 4)
```
