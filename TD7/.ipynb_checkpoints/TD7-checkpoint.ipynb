```python
import numpy as np
import matplotlib.pyplot as plt

from tqdm.auto import tqdm
```

We apply simulated annealing to obtain the maximum of the modified V-shaped distribution:

$$
\pi_k \propto
\begin{cases}
0 &\text{ if } k = 1 \\
\frac{4}{n^2} \left| \frac{n+1}2 - k\right| &\text{ otherwise}
\end{cases}
$$

for $k \in [1, n]$.

We use a logarithmic schedule:
$$ \alpha(t) = \frac 1 C \log (t + 1) $$

We index the chain from $0$ to $n+1$, with the sites at the ends of the chain being phantom sites with probability $0$.

```python
def logarithmic_schedule(C):
    """Return the logarithmic schedule for alpha with constant C"""
    def alpha_schedule(t):
        return np.log(t+1) / C
    return alpha_schedule


def modified_v_shaped_simulated_annealing(pos_ini, n_steps, n, alpha_schedule):
    """Run simulated annealing to reach the minimum of the modified V-shaped distribution. The chain length n must be even"""
    assert 1 - n%2,"n should be even"
    modified_v_shaped_distribution = [0, 0] + [abs((n + 1)/2 - k) for k in range(2, n+1)] + [0]
    pos_list = [pos_ini]
    pos = pos_ini
    for k in range(n_steps):
        alpha = alpha_schedule(k)
        x = np.random.random()
        direction = 2 * np.random.randint(2) - 1
        if x < (modified_v_shaped_distribution[pos+direction] / modified_v_shaped_distribution[pos]) ** alpha:
            pos += direction
        pos_list.append(pos)
    return pos_list
```

```python
n = 100
n_steps = 10000
C = 10
alpha_schedule = logarithmic_schedule(C)

pos_list = modified_v_shaped_simulated_annealing(n//2, n_steps, n, alpha_schedule)

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

ax[0].plot(pos_list)
ax[0].set_ylim(1, n)
ax[1].plot(alpha_schedule(np.array(range(n_steps))))
```

We compute the probability to reach $n$ for large $t$ (in principle $t \to \infty$) as a function of $C$.

```python
n = 10
n_steps = 10000
C_list = np.linspace(0.5, 5, 30)
proba_list = []

# Number of sample to compute the probability of reaching n for a given C
n_samples = 50

for C in tqdm(C_list):
    n_hits = 0
    alpha_schedule = logarithmic_schedule(C)
    for k in range(n_samples):
        if n in modified_v_shaped_simulated_annealing(n//2, n_steps, n, alpha_schedule):
            n_hits += 1
    proba_list.append(n_hits / n_samples)
```

```python
plt.plot(C_list, proba_list)
plt.xlabel("C")
plt.ylabel("Probability to reach n")
```
