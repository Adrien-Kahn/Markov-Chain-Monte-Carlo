```python
import numpy as np
import matplotlib.pyplot as plt

from tqdm.auto import tqdm
```

```python
def square_potential(x):
    return x**2 / 2

def quartic_potential(x):
    return x**4 / 4

x = np.linspace(-3, 3, 100)
plt.plot(x, square_potential(x) + quartic_potential(x))
```

```python
def factorized_MCMC(potential_1, potential_2, beta, n_steps, delta, ini_pos):
    """Performs the factorized Metropolis-Hastings algorithm for the Boltzmann distribution on the 1-dimensional line
    with the two given potential and beta"""
    pos = ini_pos
    pos_list = [pos]
    for k in tqdm(range(n_steps)):
        new_pos = pos + np.random.uniform(-delta, delta)
        metropolis_filter = min(1, np.exp(- beta * (potential_1(new_pos) - potential_1(pos)))) * min(1, np.exp(- beta * (potential_2(new_pos) - potential_2(pos))))
        if np.random.random() < metropolis_filter:
            pos = new_pos
        pos_list.append(pos)
    return pos_list
```

```python
beta = 1
n_steps = 1000000
delta = 0.1
ini_pos = 0

histogram = factorized_MCMC(square_potential, quartic_potential, beta, n_steps, delta, ini_pos)

x = np.linspace(min(histogram), max(histogram), 1000)

plt.hist(histogram, density=True, bins = 100)

exact_distribution = np.exp(- beta * (square_potential(x) + quartic_potential(x)))
exact_distribution /= exact_distribution.sum() * (x[1] - x[0])


plt.plot(x, exact_distribution)

plt.plot("Factorized MCMC")
```

```python
def factorized_MCMC_v2(potential_1, potential_2, beta, n_steps, delta, ini_pos):
    """Performs the factorized Metropolis-Hastings algorithm for the Boltzmann distribution on the 1-dimensional line
    with the two given potential and beta"""
    pos = ini_pos
    pos_list = [pos]
    for k in tqdm(range(n_steps)):
        new_pos = pos + np.random.uniform(-delta, delta)
        metropolis_filter_1 = min(1, np.exp(- beta * (potential_1(new_pos) - potential_1(pos))))
        metropolis_filter_2 = min(1, np.exp(- beta * (potential_2(new_pos) - potential_2(pos))))
        if np.random.random() < metropolis_filter_1 and np.random.random() < metropolis_filter_2:
            pos = new_pos
        pos_list.append(pos)
    return pos_list
```

```python
beta = 1
n_steps = 1000000
delta = 0.1
ini_pos = 0

histogram = factorized_MCMC_v2(square_potential, quartic_potential, beta, n_steps, delta, ini_pos)

x = np.linspace(min(histogram), max(histogram), 1000)

plt.hist(histogram, density=True, bins = 100)

exact_distribution = np.exp(- beta * (square_potential(x) + quartic_potential(x)))
exact_distribution /= exact_distribution.sum() * (x[1] - x[0])

plt.plot(x, exact_distribution)

plt.plot("Factorized MCMC version 2")
```

```python
def factorized_zig_zag(beta, n_steps, speed):
    """Performs the waiting times algorithm for the cubic and quartic potential"""
    direction = 1
    last_rejection_pos = 0
    last_rejection_time = 0
    pos_list = []
    for k in tqdm(range(n_steps)):
        new_rejection_pos_1 = (2 * np.random.exponential() / beta)**(1/2)
        new_rejection_pos_2 = (4 * np.random.exponential() / beta)**(1/4)
        new_rejection_pos = direction * min(new_rejection_pos_1, new_rejection_pos_2)
        new_rejection_time = last_rejection_time + (new_rejection_pos - last_rejection_pos) / speed
        pos_list += [(new_rejection_pos - last_rejection_pos) * (time - last_rejection_time) / (new_rejection_time - last_rejection_time) + last_rejection_pos for time in range(int(last_rejection_time), int(new_rejection_time)+1)]
        last_rejection_pos = new_rejection_pos
        last_rejection_time = new_rejection_time
        direction *= -1
    return pos_list
```
