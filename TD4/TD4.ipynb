```python
import numpy as np
import matplotlib.pyplot as plt

import itertools
from math import factorial

from tqdm.auto import tqdm
```

### Perfect sampling of permutations with top to random shuffle

```python
def top_to_random(permutation):
    """Take the left-most element of the permutation and places it at a random position in place"""
    top = permutation.pop(0)
    permutation.insert(np.random.randint(0, len(permutation)+1), top)

def top_to_random_stop(N):
    """Sample uniformly the permutations of N elements"""
    permutation = list(range(N))
    while permutation[0] != N-1:
        top_to_random(permutation)
    top_to_random(permutation)
    return permutation
```

```python
n_samples = 1000000
N = 5

histogram = {permutation: 0 for permutation in itertools.permutations(range(N))}

for k in tqdm(range(n_samples)):
    histogram[tuple(top_to_random_stop(N))] += 1
```

```python
histogram_list = [histogram[permutation] / n_samples for permutation in histogram]
plt.plot(histogram_list)
plt.ylim(0, 1.1*max(histogram_list))
plt.axhline(1 / factorial(N), c='k', ls='--')
plt.title("Distribution of {} samples".format(n_samples))
plt.xlabel("Permutation index")
plt.ylabel("Probability")
print()
```

### Perfect sampling of a particle on a line

```python
def backward_sampling(N):
    """Sample uniformly a particle on a line using perfect sampling"""
    
    current_time = 0
    all_arrows = {}
    dont_stop = True
    
    while dont_stop:
        
        # Compute the arrows at the current time
        current_arrows = [2 * np.random.randint(2) - 1 for k in range(N)]
        if current_arrows[0] == -1: current_arrows[0] = 0
        if current_arrows[-1] == 1: current_arrows[-1] = 0
        all_arrows[current_time] = current_arrows
        
        # Check if all states at the current time are mapped to a single state at time 0
        states = set(range(N))
        for time in range(current_time, 0):
            states = set([state + all_arrows[time][state] for state in states])
        dont_stop = not len(states) == 1
        
        current_time -= 1
        
    return states.pop()
```

```python
N = 10
n_samples = 1000
l = [backward_sampling(N) for k in tqdm(range(n_samples))]
plt.hist(l, bins=np.arange(-0.5, N, 0.5), density=True)
plt.title("Histogram for {} samples".format(n_samples))
print()
```
